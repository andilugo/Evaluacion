```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **California_EUA**

```python
eua = pd.read_csv("California_EUA_clean.csv")
eua
```

## Filtro a) host_acceptance_rate> 50%

```python
filtroA = eua.loc[eua['host_acceptance_rate'] > 0.50]
filtroA
```

## Filtro b) Sólo los registros con categoría “superhost”

```python
filtroB = eua.loc[eua['host_is_superhost'] == 't']
filtroB
```

## Filtro c) Los registros que no hallan verificado identidad “ not identity_verified”

```python
filtroC = eua.loc[eua['host_identity_verified'] == 'f']
filtroC
```

## Filtro d) Los registro cuyo property_type = “Private room” y “Hotel room”

```python
filtroD= eua.loc[eua['property_type'].isin(['Private room', 'Hotel room'])]
filtroD
```

## Filtro e) Los registros que cuenten con bathroom > 1

```python
filtroE = eua.loc[eua['bathrooms'] > 1]
filtroE
```

## Filtro f) Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”





```python
filtroF = eua.loc[(eua['price'] > 10000) & (eua['property_type'] == 'Entire home')]
filtroF
```

# Filtro g) Los registros cuyo review_scores_cleanliness > 4.5

```python
filtroG= eua.loc[eua['review_scores_cleanliness'] > 4.5]
filtroG
```

## Filtro h) Los registros cuyo review_scores_value > 4.9

```python
filtroH = eua.loc[eua['review_scores_value'] > 4.9]
filtroH
```

## Filtro i) Los registros cuya availability_365< 100

```python
filtroI = eua.loc[eua['availability_365'] < 100]
filtroI
```

## Filtro j) Los registros cuya host_response_time sea “within an hour”

```python
filtroJ = eua.loc[eua['host_response_time'] == 'within an hour']
filtroJ
```

```python
#Convertir DataFrame a CSV
filtroA.to_csv("FiltroA_eua.csv")
filtroB.to_csv("FiltroB_eua.csv")
filtroC.to_csv("FiltroC_eua.csv")
filtroD.to_csv("FiltroD_eua.csv")
filtroE.to_csv("FiltroE_eua.csv")
filtroF.to_csv("FiltroF_eua.csv")
filtroG.to_csv("FiltroG_eua.csv")
filtroH.to_csv("FiltroH_eua.csv")
filtroI.to_csv("FiltroI_eua.csv")
filtroJ.to_csv("FiltroJ_eua.csv")

#descargar archivo filtrado en csv
from google.colab import files 
  
files.download("FiltroA_eua.csv")
files.download("FiltroB_eua.csv")
files.download("FiltroC_eua.csv")
files.download("FiltroD_eua.csv")
files.download("FiltroE_eua.csv")
files.download("FiltroF_eua.csv")
files.download("FiltroG_eua.csv")
files.download("FiltroH_eua.csv")
files.download("FiltroI_eua.csv")
files.download("FiltroJ_eua.csv")
```
