```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **DF_México**

```python
mex = pd.read_csv("DF_México_clean.csv")
mex
```

## Filtro a) host_acceptance_rate> 50%

```python
filtroA = mex.loc[mex['host_acceptance_rate'] > 0.50]
filtroA
```

## Filtro b) Sólo los registros con categoría “superhost”

```python
filtroB = mex.loc[mex['host_is_superhost'] == 't']
filtroB
```

## Filtro c) Los registros que no hallan verificado identidad “ not identity_verified”

```python
filtroC = mex.loc[mex['host_identity_verified'] == 'f']
filtroC
```

## Filtro d) Los registro cuyo property_type = “Private room” y “Hotel room”

```python
filtroD= mex.loc[mex['property_type'].isin(['Private room', 'Hotel room'])]
filtroD
```

## Filtro e) Los registros que cuenten con bathroom > 1

```python
filtroE = mex.loc[mex['bathrooms'] > 1]
filtroE
```

## Filtro f) Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”





```python
filtroF = mex.loc[(mex['price'] > 10000) & (mex['property_type'] == 'Entire home')]
filtroF
```

# Filtro g) Los registros cuyo review_scores_cleanliness > 4.5

```python
filtroG= mex.loc[mex['review_scores_cleanliness'] > 4.5]
filtroG
```

## Filtro h) Los registros cuyo review_scores_value > 4.9

```python
filtroH = mex.loc[mex['review_scores_value'] > 4.9]
filtroH
```

## Filtro i) Los registros cuya availability_365< 100

```python
filtroI = mex.loc[mex['availability_365'] < 100]
filtroI
```

## Filtro j) Los registros cuya host_response_time sea “within an hour”

```python
filtroJ = mex.loc[mex['host_response_time'] == 'within an hour']
filtroJ
```

```python
#Convertir DataFrame a CSV
filtroA.to_csv("FiltroA_mex.csv")
filtroB.to_csv("FiltroB_mex.csv")
filtroC.to_csv("FiltroC_mex.csv")
filtroD.to_csv("FiltroD_mex.csv")
filtroE.to_csv("FiltroE_mex.csv")
filtroF.to_csv("FiltroF_mex.csv")
filtroG.to_csv("FiltroG_mex.csv")
filtroH.to_csv("FiltroH_mex.csv")
filtroI.to_csv("FiltroI_mex.csv")
filtroJ.to_csv("FiltroJ_mex.csv")

#descargar archivo filtrado en csv
from google.colab import files 
  
files.download("FiltroA_mex.csv")
files.download("FiltroB_mex.csv")
files.download("FiltroC_mex.csv")
files.download("FiltroD_mex.csv")
files.download("FiltroE_mex.csv")
files.download("FiltroF_mex.csv")
files.download("FiltroG_mex.csv")
files.download("FiltroH_mex.csv")
files.download("FiltroI_mex.csv")
files.download("FiltroJ_mex.csv")
```
