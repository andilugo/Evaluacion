```python
#datos cualitativos
esp["description"] = esp["description"].fillna("--") 
esp["neiborhood_overview"] = esp["neighborhood_overview"].fillna("--")
esp["host_name"] = esp["host_name"].fillna("--")
esp["host_since"] = esp["host_since"].fillna("--")
esp["host_location"] = esp["host_location"].fillna("--")
esp["host_about"] = esp["host_about"].fillna("--")
esp["host_response_time"] = esp["host_response_time"].fillna("--")
esp["host_is_superhost"] = esp["host_is_superhost"].fillna("--")
esp["host_thumbnail_url"] = esp["host_thumbnail_url"].fillna("--")
esp["host_picture_url"] = esp["host_picture_url"].fillna("--")
esp["host_thumbnail_url"] = esp["host_thumbnail_url"].fillna("--")
esp["host_has_profile_pic"] = esp["host_has_profile_pic"].fillna("--")
esp["host_identity_verified"] = esp["host_identity_verified"].fillna("--")
esp["neighbourhood"] = esp["neighbourhood"].fillna("--")
esp["bathrooms_text"] = esp["bathrooms_text"].fillna("--")
esp["first_review"] = esp["first_review"].fillna("--")
esp["last_review"] = esp["last_review"].fillna("--")
esp["license"] = esp["license"].fillna("--")
esp["host_response_rate"] = esp["host_response_rate"].fillna("0%")
esp["host_acceptance_rate"] = esp["host_acceptance_rate"].fillna("0%")
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=esp_num
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers del Dataframe
outliers= esp_num[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
esp2 = esp_num[(y<=Limite_Superior)&(y>=Limite_Inferior)]
esp2
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=esp_num.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=esp_num.copy()
data_clean=data_clean.fillna(round(esp_num.mean(),1))
data_clean
```
